```
from sklearn.preprocessing import MinMaxScaler
```

```
X=X.values
```

```
y=y.values
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```
myScaler=MinMaxScaler()
```

```
X_train=myScaler.fit_transform(X_train)
```

```
X_test=myScaler.transform(X_test)
```

```
X_train.shape
```

```
from tensorflow.keras.models import Sequential
```

```
from tensorflow.keras.layers import Dense,Dropout
```

```
myTFmodel=Sequential()
```

```
myTFmodel.add(Dense (25,'relu'))
myTFmodel.add(Dropout(0.2))

myTFmodel.add(Dense (13,'relu'))
myTFmodel.add(Dropout(0.2))

myTFmodel.add(Dense (7,'relu'))
myTFmodel.add(Dropout(0.2))

myTFmodel.add(Dense (1,'linear'))
```

```
myTFmodel.compile(optimizer='adam', loss='mean_squared_error')
```

```
myTFmodel.fit(X_train,y_train,epochs=200,validation_split = 0.2)
```

```
plt.plot(pd.DataFrame(myTFmodel.history.history))
```
